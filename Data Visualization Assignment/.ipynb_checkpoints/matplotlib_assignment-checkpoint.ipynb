# Problem 1

```python
# Importing libraries
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import optimize
```

```python
temp_max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
temp_min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])

months = np.arange(12)
```

```python
# Fitting to the periodic function
def yearly_temps(times, avg, ampl, time_offset):
    return (avg+ ampl * np.cos((times + time_offset) * 1.8 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, months,
                                      temp_max, [40, 20, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months,
                                      temp_min, [-40, 20, 0])
```

```python
days = np.linspace(0, 12, num=365)

plt.figure()
plt.plot(months, temp_max, 'go')
plt.plot(days, yearly_temps(days, *res_max), 'm-')
plt.plot(months, temp_min, 'co')
plt.plot(days, yearly_temps(days, *res_min), 'y-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')
plt.show()
```

# Problem 2

```python
import pandas as pd
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv")
data.head()
```

### Pie chart presenting male/female proportion

```python
data['sex'].value_counts()
```

```python
gender = list(data['sex'].value_counts().index)
gender_proportion = list(data['sex'].value_counts().values)
print(gender)
print()
print(gender_proportion)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
ax.pie(gender_proportion, labels = gender, autopct="%1.2f%%")
plt.title("Gender Proportion")
plt.show()
```

```python
x = data['fare'].values
y = data['age'].values
c = list(data['sex'].value_counts().index)

fig, ax = plt.subplots()
for label,color in zip(c, ["blue", 'red']):
    ax.scatter(x, y, c=color, label=label,
               alpha=0.3, edgecolors='none')

ax.legend()
ax.grid(True)

plt.show()
```
